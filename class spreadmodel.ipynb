```python
years = ['2014','2015','2016','2017']
rolling = [1,2,3,4,5,6,7]
results = []
lr = LogisticRegression()
def rolling_sampler(years,rolling,without_game_1=False):
    results = []
    for year in years:
        for number in rolling:
            train = SpreadModel(year=year,rolling_avg=number,without_game_1=without_game_1)
            test = SpreadModel(year=str(int(year)+1),rolling_avg=number,without_game_1=without_game_1)
            X_train, y_train = train.create_model_data()
            X_test, y_test = test.create_model_data()
            lr.fit(X_train,y_train.values.reshape(-1))
            score = lr.score(X_test,y_test.values.reshape(-1))
            print(f'Test Year: {str(int(year)+1)} Rolling Avg: {number} Score:{score}')
            results.append([(year,number,score)])
    return results
```
